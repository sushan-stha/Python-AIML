<p><b>10. Use of reshape</b></p>
<sub><span style="color: pink">First creating a simple list of numbers from 0 to 23</span></sub><br>


```python
g=np.arange(24)
print(g)
print("rank:",g.ndim)
```

<sub><span style="color: pink">Then changing dimension of array without changing actual data elements</span></sub>

```python
g2=g.reshape(4,6)
print(g2)
print("rank:",g2.ndim)
```

<sub><span style="color: pink">Creating another 2x2 list </span></sub>

```python
a=np.zeros((2,2))
print(a)
```

<sub><span style="color: pink">.reshape(1, 4): Flattens those 4 elements into a single horizontal row containing 4 columns.</span></sub>

```python
g3=a.reshape(1,4)
print(g3)
print("rank:",g3.ndim)
```

<p><b>12. Flattening Arrays: ravel(), flatten() & reshape()</b></p>

<sub><span style="color: pink">Used to collapse multi-dimensional arrays into flat, simple 1D array</span></sub>

<sub><span style="color: green"><b>.ravel()</b>:<br>
points directly to your original data<br>
changes original data too<br>
Very fast and no use of extra memory</span></sub>

```python
a.ravel()
```

<sub><span style="color: green"><b>.flatten()</b>:<br>
makes a brand-new, independent copy of your data in memory<br>
Changing values here will not affect your original array</span></sub>

```python
a.flatten()
```

```python
b=np.zeros((100,100))
print(b)
```

```python
b=b.ravel()
```

```python
b=b.flatten()
```

<sub><span style="color: green"><b>.reshape(-1)</b>:<br>
automatically flat-pack everthing into 1D array<br>
Like ravel(), this is just a shortcut (view) of the original array, not a copy.</span></sub>

```python
b=b.reshape(-1)
```
