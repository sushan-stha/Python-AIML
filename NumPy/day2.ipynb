#rules of broadcasting|

#First Rule

```python
import numpy as np
h = np.arange(5).reshape(1, 1, 5)
```

```python
h + [10, 20, 30, 40, 50]
```

#Second Rule

```python
k = np.arange(6).reshape(2, 3)
```

```python
k+[[100], [200]]
```

#Third rule


```python
k = np.arange(9).reshape(3, 3)
try:
    k + [33, 44]
except ValueError as e:
    print(e)
```
