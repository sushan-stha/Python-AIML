#First Rule

```python
import numpy as np
h = np.arange(5).reshape(1, 1, 5)
```

```python
h + [10, 20, 30, 40, 50]
```

#Second Rule

```python
k = np.arange(6).reshape(2, 3)
```

```python
k+[[100], [200]]
```

#Third rule


```python
k = np.arange(9).reshape(3, 3)
try:
    k + [33, 44]
except ValueError as e:
    print(e)
```

```python
C = A.reshape(3, 1) + B
print(C)
```

```python
answer = A[:,np.newaxis]+B
print(answer)
```

<h2>Upcasting</h2>
<sub>-when trying to combine arrays with different dtypes, NumPy will upcast to a type capable of handling all possible values</sub>

```python
k1 = np.arange(0, 5, dtype=np.uint8)
```

```python
k2 = k1 + np.array([5, 6, 7, 8, 9], dtype=np.int8)
print(k2.dtype, k2)
```

#Broadcasting with conditional operators

#Mathematical and Statistical Functions<br>
<sub>~This performs the operation on all elements in the ndarray, regardless of its shape</sub>

```python
a = np.array([[-.25, 3.1, 7], [10, 11, 12]])
```

```python
for func in (a.min, a.max, a.sum, a.prod, a.std, a.var, a.mean):
    print(func,__name__,"=", func())
```
